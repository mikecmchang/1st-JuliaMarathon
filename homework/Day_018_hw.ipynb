# DataFrames.jl 介紹 (二): Joins 與 Split-Apply-Combine Strategy

## Day 018 作業1：The Split-Apply-Combine Strategy

同 Day 017，請載入 2020/4/2 COVID-19 Daily Report 資料集，計算每個國家/地區的累計確診人數，並列出所有資料筆數。

```julia
using DataFrames, CSV
```

```julia
df = CSV.read("Day_017_hw_COVID-19.04-02-2020.csv", delim = ',') # 撰寫讀取 CSV 
```

計算計算每個國家/地區的累計確診人數

【提示】可使用 groupby 相關的函式

```julia
df_confirmed = by(df, :Country_Region, Total_Confirmed = :Confirmed => sum);
df_confirmed = sort(df_confirmed, :Total_Confirmed, rev = true);
```

```julia
# 顯示所有資料
show(df_confirmed, allrows = true, allcols = true)
```

## 作業2：請列出美國各區域或事件 (依 Province_State 欄) 的累積確診、死亡、康復人數。

【提示】使用 Split-Apply-Combine Strategy 時，有幾種不同的函式可以使用，請參照 `DataFrames.jl` 官方文件 [https://juliadata.github.io/DataFrames.jl/stable/lib/functions/#DataFrames.groupby](https://juliadata.github.io/DataFrames.jl/stable/lib/functions/#DataFrames.groupby)

```julia
using DataFrames, CSV
df = CSV.read("Day_017_hw_COVID-19.04-02-2020.csv", delim = ',')
df_us = df[df.Country_Region .== "US", :];
df_us_stat = by(df_us, :Province_State, Total_Confirmed = :Confirmed => sum, Total_Deaths = :Deaths => sum, Total_Recovered = :Recovered => sum);
df_us_stat = sort(df_us_stat, :Total_Confirmed, rev = true);
show(df_us_stat, allrows = true, allcols = true)
```
